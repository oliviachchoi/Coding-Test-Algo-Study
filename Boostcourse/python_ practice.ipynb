```python
#실습문제

def computepay(hours, rate):
    if hours > 40:
        reg = rate * hours
        otp = (hours - 40.0) * (rate * 0.5)
        pay = reg + otp
    else:
        pay = hours * rate
    return pay

sh = input("Enter Hours: ")
sr = input("Enter Rate: ")
fh = float(sh)
fr = float(sr)

xp = computepay(fh, fr)
#fh = hours, fr = rate

print("Pay: ", xp)
```

##while

```python
#While, Break
while True:
  line= input('>')
  if line== 'done':
    break #breaking out of a loop
  print(line)
print('Done!')  
```

```python
#while continue
#continue는 루프문의 첫 줄로 다시 돌아감
#break는 그 다음으로 넘어감

#in this case, we used 'continue', therefore it will not activate the next if sentence
# #로 시작하는 단어는 print문 무시.
while True:
  line= input('> ')
  if line[0]=='#':
    continue
  if line =='done':
    break
  print(line)
print('Done!')
```

```python
counts=dict()
names = ['csev', 'cwen', 'csev', 'zqian', 'cwen']

for name in names:
  if name in counts:
    counts[name]= counts[name]+1
  else:
    counts[name]=1
print(counts)
```

```python
counts=dict()
names = ['csev', 'cwen', 'csev', 'zqian', 'cwen']
for name in names:
  if name not in counts: #not in
    counts[name]=1
  else: 
    counts[name]= counts[name]+1
print(counts)
```

```python
#get 함수 사용
counts={}
names = ['csev', 'cwen', 'csev', 'zqian', 'cwen']
for name in names:
  counts[name]= counts.get(name,0)+1
print(counts)
```
